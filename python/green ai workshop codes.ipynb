```python
# Average carbon footprint calculation
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]

total_carbon = 0

# Calculate the total carbon footprint
for city in climate_data:
    total_carbon += city["carbon_footprint"]

# Calculate the average carbon footprint
average_carbon_footprint = total_carbon / len(climate_data)

# Print the result with 2 decimal places
print(f"\nAverage Carbon Footprint: {average_carbon_footprint:.2f}")
```

```python
def calculate_carbon_footprint(energy_consumption, emission_factor):
    """
    this function calculates the carbon footprint based on energy consumptipn
    (in kilowatt-hours) and the emission factor(in kg per kkh).
    """
    return energy_consumption * emission_factor
```

```python
energy_consumption = 1000 #in kWh
emission_factor = 0.475 #kg co2 per kwh

carbon_footprint = calculate_carbon_footprint(energy_consumption, emission_factor)
print(f"Carbon Footprint: {carbon_footprint} kg CO2")
```

```python
import pandas as pd
import numpy as np

data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}

#create a dataframe
energy_df = pd.DataFrame(data)
print("Original energy data with missing values")
energy_df.head() #for first five rows
```

```python
energy_df.tail() #last five rows
```

```python
#remove rows with any missing values
cleaned_df = energy_df.dropna()
print("\nData after removing rows with missing values: ")
cleaned_df.head()
```

```python
#print 2nd and 3rd element using ilog function
print(energy_df.iloc[[1,2]])
```

```python
ec_mean=energy_df["Energy Consumption (MWh)"].mean()
ec_mean
print("Mean of Energy consumption (MWh)",ec_mean)
cost_mean = energy_df["Cost (Million $)"].mean()
print("Mean of Cost (Million $)" , cost_mean)
```

```python
#input missing values in energy consumption with mean
energy_df["Energy Consumption (MWh)"].fillna(ec_mean,inplace=True)

energy_df["Cost (Million $)"] = energy_df["Cost (Million $)"].fillna(cost_mean)

print("\nDataFrame with Imputed Values:")
energy_df.head()
```

```python
#forward filling
forward_filled_df = energy_df.fillna(method = "ffill")
print("\nData before forward filling:")
print(energy_df)
print("\nData after forward filling:")
forward_filled_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

#normalize
scaler = MinMaxScaler()
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(
    energy_df[["Energy Consumption (MWh)" , "Cost (Million $)"]]
)
print("\nData after normalization (Min-Max Scaling):")
print(energy_df)
```

```python
#standard scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(
    energy_df[["Energy Consumption (MWh)" , "Cost (Million $)"]]
)
print("\nData after standardization (Z - score Scaling):")
print(energy_df)
```

```python
#encoding categorical variables 
#one hot encode the energy source column
energy_encoded_df = pd.get_dummies(energy_df, columns=["Energy Source"])

print("\nData after one-hot encoding categorical variables:")
energy_encoded_df.head()
```

```python
#pairplot - Visualizing Relationships Between Renewable Energy Sources and Their Consumption
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame for pair plot
data = {
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df = pd.DataFrame(data)
# Create a pair plot
sns.pairplot(df)
plt.show()
```

```python
#heatmap - Showing the Correlation Between Different Types of Energy Consumption
# Create a heatmap to show the correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Between Energy Sources')
plt.show()
```

```python
#Box Plot: Comparing the Distribution of Energy Consumption Across Different Sources
# Create a box plot
sns.boxplot(data=df)

# Add title
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
# Violin Plot: Comparing Energy Consumption Distributions
# Create a violin plot
sns.violinplot(data=df)

# Add title
plt.title('Violin Plot of Energy Consumption Distributions')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Regression Plot: Analyzing the Relationship Between Energy Consumption and Carbon Emissions
# Sample data for energy consumption and carbon emissions
energy_values = [100, 200, 300, 400, 500]  # Replace with your actual data
carbon_emissions = [10, 20, 25, 40, 50]   # Replace with your actual data

# Create a DataFrame with energy consumption and carbon emissions
df_reg = pd.DataFrame({
    'Energy Consumption': energy_values,
    'Carbon Emissions': carbon_emissions
})

# Create a regression plot
sns.regplot(x='Energy Consumption', y='Carbon Emissions', data=df_reg)

# Add title
plt.title('Energy Consumption vs Carbon Emissions')
plt.xlabel('Energy Consumption')
plt.ylabel('Carbon Emissions')
plt.show()
```

```python
# Facet Grid: Plotting Energy Consumption Over Time for Different Regions
# Define the months
months = ['January', 'February', 'March', 'April', 'May', 'June']

# Creating a dataset for different regions and energy consumption over months
df_facet = pd.DataFrame({
    'Month': months * 3,
    'Energy Consumption': [1200, 1300, 1100, 1500, 1400, 1600] * 3,
    'Region': ['North', 'South', 'East'] * 6
})

# Create a FacetGrid plot
g = sns.FacetGrid(df_facet, col='Region', hue='Region', height=4, aspect=1)
g.map(sns.lineplot, 'Month', 'Energy Consumption')

# Add title
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Energy Consumption Over Time by Region')
plt.show()
```
